Part 3

```python
dsf=pd.read_csv('https://raw.githubusercontent.com/aaiken1/fin-data-analysis-python/main/data/crsp_022722.csv',parse_dates=[('date')])
dsf.info()

#parse_dates doesn't always work. If this problem occurs, use pd.to_datetime
```

```python
dsf
```

```python
#ticker is a col, hence it is a long data
```

```python
dsf['PRC_ADJ']=dsf.PRC/(dsf.CFACPR)
dsf
```

```python
dsf['PRC_ADJ_LAG1']=dsf.groupby(['PERMNO'])['PRC_ADJ'].shift(1)
dsf
```

```python
dsf['DIVAMT']=dsf.DIVAMT.fillna(0)
```

```python
dsf['RETURN']=(((dsf.PRC_ADJ+dsf.DIVAMT)/(dsf.PRC_ADJ_LAG1))-1)
dsf
```

```python
dsf.RET-dsf.RETURN
```

Part 5

```python
#gross return variable
dsf['RET_G']=dsf.RET+1
dsf
```

```python
dsf['Cum_Ret']=dsf.groupby(['PERMNO'])['RET_G'].cumprod().sub(1)
dsf
```

Part 6

```python
dsf.set_index('date').inplace=True
```

```python
dsf.groupby('PERMNO')['Cum_Ret'].plot(legend=True, title = 'Cumulative Returns');
```

```python
#Reset index
dsf.reset_index()
```

```python
dsf_wide = dsf.pivot_table(index='date', columns='PERMNO', values='Cum_Ret')
dsf_wide
```

```python
dsf_wide.plot(subplots = True)
```

Part 7

```python
dsf_ret_wide = dsf.pivot_table(index='date', columns='PERMNO', values='RET')
dsf_ret_wide
```

```python
dsf_price_wide = dsf.pivot_table(index='date', columns='TICKER', values='PRC_ADJ')
dsf_price_wide
```

Part 8

```python
#Create a simple 30-day moving average for the price of each security
dsf_price_wide['AAPL_PRC_30']=dsf_price_wide['AAPL'].rolling(window=30).mean()
dsf_price_wide['TLT_PRC_30']=dsf_price_wide['TLT'].rolling(window=30).mean()
dsf_price_wide['XOM_PRC_30']=dsf_price_wide['XOM'].rolling(window=30).mean()
```

```python
fig, axs = plt.subplots(2, 1, sharex=True, sharey=False,figsize=(10, 6))

axs[0].plot(dsf_price_wide['AAPL'] , label = 'AAPL_PRICE')
axs[0].plot(dsf_price_wide['AAPL_PRC_30'] , label = 'AAPL_PRC_30')
axs[0].set_title('APPL')
axs[0].legend()

axs[1].plot(dsf_price_wide['TLT'],label='TLT_PRC')
axs[1].plot(dsf_price_wide['TLT_PRC_30'],label='TLL_PRC_30')
axs[1].set_title('TLT')
axs[1].legend()



#axs[2].plot(dsf_price_wide, 'b', label = 'BTC Daily Return')
```
